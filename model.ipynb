3 neuron model options
1. Hodgkin-Huxley: higher accuracy, heavy
2. Artificial: simplified, activation_func(inputs*weights)
3. Leaky Integrate-and-Fire: if integrate(inputs*weights) > threshold, then neuron emits voltage spike -> spiking time/freq matters, spiking value doesn't

# **4. LIF**

We choose LIF... Imagine a **bucket** (neuron model) that collects **water drops** (signals), but has a **small hole** at the bottom (leak) causing the water level to go down over time. Once the **water reaches a certain level** (threshold), it **dumps all the water** (fires) and starts empty again, the firing is a discrete event.

<font color="red">τ</font>&nbsp;&nbsp;&nbsp;&nbsp;: leak rate constant = <font color="lightblue">R</font>*C \\
